```python
#Se muestra el DataFrame creado a partir del documento xlsx
p = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
productos = pd.DataFrame(p) 
productos
```

```python
productos.info()
```

```python
productos.head()
```

```python
#Identificar los valores nulos por columna 
valores_nulos= productos.isnull().sum()
valores_nulos
```

```python
#Identificamos que solo existen dos valores nulos y los cambiamos por "--" al ser valores unicos 
productos["NOMBRE_VENDEDOR"]=productos["NOMBRE_VENDEDOR"].fillna("--")
```

```python
#Verificamos que no haya  valores nulos 
valores_nulos= productos.isnull().sum()
valores_nulos
```

```python
#Dada la relevancia de los datos eliminamos algunas columnas 
# Eliminamos las columnas "NOMBRE_VENDEDOR", "DESCR", "COSTO_UNITARIO", "SUBTOTAL_PARTIDA", "MARGEN_UNITARIO_CALCULADO" 
productos = productos.drop(['NOMBRE_VENDEDOR','DESCR', 'COSTO_UNITARIO', 'SUBTOTAL_PARTIDA', 'MARGEN_UNITARIO_CALCULADO'], axis=1)

# Mostramos el DataFrame modificado
print('DataFrame modificado:')
productos
```

```python
#Convertir DataFrame a CSV 
productos.to_csv("detalleprecio_sin_nulos.csv")
```
